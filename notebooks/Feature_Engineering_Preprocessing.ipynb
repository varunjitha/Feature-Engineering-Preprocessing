```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("../data/raw/Titanic-Dataset.csv")

df.head()
```

```python
df.info()

df.describe()

df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())

df["Embarked"] = df["Embarked"].fillna(
    df["Embarked"].mode()[0]
)

df["Fare"] = df["Fare"].fillna(
    df["Fare"].median()
)
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
```

```python
df["IsAlone"] = np.where(
    df["FamilySize"] == 1,
    1,
    0
)
```

```python
df["Title"] = df["Name"].str.extract(
    r"([A-Za-z]+)\.",
    expand=False
)
```

```python
df.drop(
    ["PassengerId", "Name", "Ticket", "Cabin"],
    axis=1,
    inplace=True
)
```

```python
le = LabelEncoder()

df["Sex"] = le.fit_transform(df["Sex"])
```

```python
df = pd.get_dummies(
    df,
    columns=["Embarked", "Title"],
    drop_first=True
)
```

```python
scaler = StandardScaler()

df[["Age", "Fare"]] = scaler.fit_transform(
    df[["Age", "Fare"]]
)
```

```python
minmax = MinMaxScaler()

df[["FamilySize"]] = minmax.fit_transform(
    df[["FamilySize"]]
)
```

```python
df.to_csv(
    "../data/processed/Titanic_Preprocessed.csv",
    index=False
)
```
